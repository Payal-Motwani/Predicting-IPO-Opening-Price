```python

df1 = df1.dropna(axis=0)
df1.isna().sum()
df1.dtypes
```

```python
'''Data cleaning - Column types'''
df1.columns = priced_headers
'''Transforming date to proper format'''
df1['Date']= pd.to_datetime(df1['Date'])
```

```python
'''Converting Price from str to float'''
df1['Offer Amount'] =  df1['Offer Amount'].map(lambda x: x.replace("$", "").replace(",", ""))
df1['Offer Amount'] = pd.to_numeric(df1['Offer Amount'], errors='coerce')
```

```python
df1['Price'] = df1['Price'].map(lambda x: x.replace("$", ""))
df1['Price'] = pd.to_numeric(df1['Price'], errors='coerce')
```

```python

df1['Shares'] = df1['Shares'].str.replace(',','')
df1['Shares'] = pd.to_numeric(df1['Shares'], errors='coerce')
```

```python
'''On inspecting company names, certain ETFs and trust were found, as I am inspecting , I will be eliminating these values'''
'''I am interested in those companies that add value or in other words are truly new companies. 
However, there are also so called "special purpose acquisition companies" that go to IPO. I'm not interested in them.

So we want to exclude companies: that have "merge" | "acquisition" in their names'''
def clean_data(data):
    cond1 = data['Company Name'].str.contains('ACQUISITION|MERGER', case=False, regex=True)
    cond2 = data['Company Name'].str.contains(' ETF', case=False, regex=True)
    cond3 = data['Company Name'].str.contains('TRUST', case=False, regex=True)
    return data [~cond1 & ~cond2 & ~cond3]
```

```python
clean_data =clean_data(df1)
```
